```python
# 정규화 함수
def norm_1(context):
    return emoticon_normalize(context, num_repeats=2)

def norm_2(context):
    hangul = re.compile('[^가-힣]')
    result = hangul.sub(' ', str(context))
    #result = result.replace(' ','')
    return result



# 띄어쓰기 함수
def space(context):
    # 띄어쓰기 없애고
    context = context.replace(' ', '')
    # 띄어쓰기 추가
    try:
      result = spacing(context)
      return result
    
    except:
      return context



# # 맞춤법 함수
# def hangle(context):
#     try:
#       a = spell_checker.check(context)
#       result = a.checked
#       return result

#     except:
#       return context




# 불용어 함수
# stopwords = set(stopwords.split(' '))

# def del_data(context):
#   result = [word for word in context if not word in stopwords]
#   return result 


# okt 토큰화 함수
def okt_token(context):
  return okt.morphs(context)


# customized 토큰화 함수
def custome_token(context):
  return twitter.morphs(context)


# L token 토큰화 함수
scores = {word:score.cohesion_forward for word, score in word_score_table.items()}
l_tokenizer = LTokenizer(scores=scores)

def l_token(context):
  return l_tokenizer.tokenize(context, flatten=False)



불용어 = []
# 리스트를 딕셔너리로 생성하고 빈도수에 따라 키워드를 정렬하는 함수
def count_f(my_list):
    count = {} # 딕셔너리 생성

    for item in my_list:
        count[item] = count.get(item, 0) + 1
    
    final_df = (pd.DataFrame(list(count.items()),
                            columns=['keyword', 'count']))
    
    # 빈도수가 10보다 작다면 해당 행 삭제
    d_row = final_df[final_df['count'] < 10].index
    불용어.append(final_df[final_df['count'] < 10])
    final_df = final_df.drop(d_row)

    # count 칼럼에 따라서 내림차순 정렬
    final_df = final_df.sort_values(by=['count', 'keyword'], ascending=False)
    final_df = final_df.reset_index(drop=True) # index 초기화    
    
    return final_df


  

# 데이터프레임 csv파일로 저장하는 함수
def df_export_csv(my_df):
    my_file = input('저장할 파일의 이름을 입력하세요 : ')
    exp = my_df.to_csv(('{0}.csv').format(my_file), index=False, header=False, encoding='cp949')
    
    return exp
  
```

```python
def change(context):
    if context[0][0] in ['즣', '좋', '조']:
        a = re.sub('즣아+[ㄱ-ㅎ가-힣]+|좋내여+[가-힣]+|좋다+[가-힣]+|조아+[가-힣]+|좋았+[가-힣]+', '좋아요', context)
        return a
    
    if context[0][0] == '따':
        a = re.sub('따갑|따갑+[ㄱ-ㅎ가-힣]+|따가+[가-힣]+|따가+[가-힣]+', '따가움', context)
        return a
    
    if context[0][0] == '느':
        a = re.sub('느리+[ㄱ-ㅎ가-힣]+, 느림+[ㄱ-ㅎ가-힣]+', '느려', context)
        return a
    
    if context[0][0] == '넉':
        a = re.sub('넉넉+[ㄱ-ㅎ가-힣]+', '넉넉', context)
        return a
    
    if context[0][0] == '모':
        a = re.sub('모르+[ㄱ-ㅎ가-힣]+|모름+[가-힣]+', '모름', context)
        return a
    
    if context[0][0] == '비':
        a = re.sub('비싸+[ㄱ-ㅎ가-힣]+|비싼+[가-힣]+', '비쌈', context)
        return a
    
    if context[0][0] == '시':
        a = re.sub('시려+[ㄱ-ㅎ가-힣]+|시린+[가-힣]+|시리+[가-힣]+', '시린', context)
        return a
    
    if context[0][0] in ['매', '맵']:
        a = re.sub('매워+[ㄱ-ㅎ가-힣]+|맵네+[가-힣]+|매운+[가-힣]+', '매움', context)
        return a
    
    if context[0][0] == '깔':
        a = re.sub('깔끔+[ㄱ-ㅎ가-힣]+', '깔끔', context)
        return a
    
    if context[0][0] == '깨':
        a = re.sub('깨끗+[ㄱ-ㅎ가-힣]+', '깨끗', context)
        return a
    
    if context[0][0] == '이':
        a = re.sub('이빨+[ㄱ-ㅎ가-힣]+', '치아', context)
        return a
    
    if context[0][0] == '없':
        a = re.sub('없네+[ㄱ-ㅎ가-힣]+|없어+[가-힣]+', '없음', context)
        return a
    
    if context[0][0] == '양':
        a = re.sub('양치질+[ㄱ-ㅎ가-힣]+', '양치', context)
        return a
    
    if context[0][0] == '상':
        a = re.sub('상쾌+[ㄱ-ㅎ가-힣]+', '상쾌', context)
        return a
    
    if context[0][0] == '개':
        a = re.sub('개운+[ㄱ-ㅎ가-힣]+', '개운', context)
        return a
```
