```python
import pandas as pd
import openpyxl
import datetime
import csv
```

```python
# df = pd.read_excel('./칫솔(SNG 치과칫솔).xlsx', index_col=0)
# df.to_csv('./칫솔(SNG 치과칫솔).csv')
df = pd.read_csv('./칫솔(SNG 치과칫솔).csv', index_col=0)
```

### 데이터 수치화

#### '내용' column 중 앞에 붙은 단어들 삭제

```python
# 데이터 프레임 중 첫 단어 없애기(BEST,한달사용기,\n)
df['내용'] = df['내용'].replace({'BEST':'','한달사용기':'', '\n':' '}, regex=True) 

# 데이터 프레임 중 첫 단어 없애기(특정단어가 붙어있는 문장에서도 삭제할수 있다)
for col in '내용':
    df['내용'] = df['내용'].apply(lambda x : x[3:] if x.startswith("재구매") else x) 
for col in '종류':
    df['종류'] = df['종류'].apply(lambda x : x[0:] if x.startswith(" ") else x)
df
```

#### column들 int로 전환

```python
# 종류의 숫자부분 앞뒤로 문자삭제
df['종류'] = df['종류'].str.split(':').str[-1]    # ':' 문자 이전의 문자들 삭제
df['종류'] = df['종류'].str.split('.').str[0]     # '.' 문자 이후의 문자들 삭제
df
```

```python
# 종류column 정의 
# 재구매와 한달사용기를 각각 1로 정의
df['재구매'] = df['재구매'].replace('재구매','1')
df['한달사용기'] = df['한달사용기'].replace('한달사용기','1')

# 1번: 미세모, 2번 : 이중미세모, 3번: 일반모 ,4번: 어금니전용+스파이럴모, 5번 : 치실과 치간칫솔, 6번 : 미세모 반, 일반모 반으로 주문, 7번: 이중미세모 반, 미세모 반
df['종류'] = df['종류'].str.replace('3','2')
df['종류'] = df['종류'].str.replace('4','1')
df['종류'] = df['종류'].str.replace('5','1')
df['종류'] = df['종류'].str.replace('A','3')
df['종류'] = df['종류'].str.replace('6','2') 
df['종류'] = df['종류'].str.replace('7','1')
df['종류'] = df['종류'].str.replace('8','3')
df['종류'] = df['종류'].str.replace('9','4')
df['종류'] = df['종류'].str.replace('10','4')
df['종류'] = df['종류'].str.replace('11','5')
df['종류'] = df['종류'].str.replace('12','5')
df
```

```python
# 남은 과거의 주문 데이터는 종류를 알수없음. ♥,@로만 나왔다.
for i in range(len(df)):
    if df['종류'][i].count('일반') == 1:
        df['종류'][i] = '6'
    elif df['종류'][i].count('이중슬림') == 1:
        df['종류'][i] = '7'
    elif df['종류'][i].count('곰돌이') == 1:
        df['종류'][i] = '7'
    elif df['종류'][i].count('슬림') == 2:
        df['종류'][i] = '1'
    elif df['종류'][i].count('앤드') == 2:
        df['종류'][i] = '4'
    elif df['종류'][i].count('배송') == 1:
        df['종류'][i] = '0'
df
```

```python
# string을 int화
df['재구매'] = df['재구매'].astype(int)
df['한달사용기'] = df['한달사용기'].astype(int)
df['종류'] = df['종류'].astype(int)
df.info()
df
```

#### 날짜 column을 datetime type으로 변환
날짜별로 정렬 후 인덱스도 재정렬

```python
# 마지막에 붙던 .를 모든 row에서 삭제
df['날짜'] = df['날짜'].map(lambda x: str(x)[:-1])
df
```

```python
# 2000년대를 뜻하는 20을 더함(더하지않을시 to_datetime 활용시 년도가 정확하게 나오지않았다.)
df['날짜'] ='20' + df['날짜'].astype(str)
df
```

```python
# 날짜순으로 재정렬후 index 리셋
df = df.sort_values(by=['날짜'], axis=0)
df = df.reset_index(drop=True)
df
```

```python
# 날짜를 datetime변환
df['날짜'] = pd.to_datetime(df['날짜'])
```

#### 저장

```python
df.to_csv('./SNG칫솔 전처리 완료.csv')
```

#### column들 int로 전환

```python
# 데이터에 NaN값이 있었기에 읽어내질 못함. 문자 '0'으로 결측치를 채운다.('0'은 알수없음임)
df3['종류'] = df3['종류'].fillna('0')
```

```python
# 종류column 정의 
# 재구매와 한달사용기를 각각 1로 정의
df3['재구매'] = df3['재구매'].replace('재구매','1')
df3['한달사용기'] = df3['한달사용기'].replace('한달사용기','1')

# 1번: 미세모, 2번 : 이중미세모, 3번: 일반모 ,4번 : 어금니전용+스파이럴모, 5번 : 치실과 치간칫솔, 6번 : 미세모 반, 일반모 반으로 주문, 7번: 이중미세모 반, 미세모 반 
# 0번 알수없음(미선택 + 배송메세지로 적었기에 알수없음 + 단종)
for i, item in enumerate(df3['종류']):
    if '이중' in item:              # 이중미세모 변환
        df3['종류'][i] = '2'
    elif '반반' in item:              # 미세모,일반모세트 변환
        df3['종류'][i] = '6'
    elif '치간' in item:              # 치간칫솔 변환
        df3['종류'][i] = '5'
    elif '일반' in item:              # 일반모 변환
        df3['종류'][i] = '3'
    elif '스파이럴' in item:          # 스파이럴모 변환
        df3['종류'][i] = '4'
    elif '어금니' in item:            # 어금니전용 변환
        df3['종류'][i] = '4'
    elif '미세' in item:              # 미세모 변환
        df3['종류'][i] = '1'
    elif '와타나베' in item:          # 일반모 변환
        df3['종류'][i] = '3'
    elif '단단' in item:              # 일반모 변환
        df3['종류'][i] = '3'
    elif 'E-1' in item:              #  E-1,2,3,4는 유아용제품. 연령대 + 이중미세모와 미세모로 구분된 제품군이라 각각 구분
        df3['종류'][i] = '1'
    elif 'E-2' in item:
        df3['종류'][i] = '2'
    elif 'E-3' in item:
        df3['종류'][i] = '1'
    elif 'E-4' in item:
        df3['종류'][i] = '2'
    elif '대왕' in item:              # 대왕 칫솔 실속패키지는 단종
        df3['종류'][i] = '0'
df3
```
