```python
# 1. 비식별화 이후 정보없음 -> 삭제
df = df.drop(columns = ['아이디or이메일'])

# 2. datetime 변환
# 파생변수 만들때 한꺼번에 진행
```

```python
# 3.dummy variable

df = df.replace('y',1)
df = df.replace('n',0)

df = df.replace('비회원', 0) 
df = df.replace('회원', 1) 

df = df.replace('남성', 0) 
df = df.replace('여성', 1)

df = df.replace('좁은구강구조', 0) 
df = df.replace('보통구강구조', 1)
df = df.replace('넓은구강구조', 2)

df.head(5)
```

```python
# 4. 단위제거

df['양치시간(분)'] = df['양치시간'].str.get(i=0)
df = df.drop(columns = ['양치시간'])

df['키(cm)'] = df['키'].str.replace("cm", "")
df = df.drop(columns = ['키'])

df['몸무게(kg)'] = df['몸무게'].str.replace("kg", "")
df = df.drop(columns = ['몸무게'])

df['하루양치질(회)'] = df['하루 양치질 횟수'].str.replace("회", "")
df = df.drop(columns = ['하루 양치질 횟수'])

df.head()
```

```python
# 5. tidy data 처리

df['전체회차'] = df['회차'].str.split('/').str[0]
df['진행된회차'] = df['회차'].str.split('/').str[1]

df['추천칫솔'] = df['추천결과'].str.split('/').str[0]
df['추천치약'] = df['추천결과'].str.split('/').str[1]

df = df.drop(columns = ['회차'])
df = df.drop(columns = ['추천결과'])

df.head()
```

```python
# 데이터 타입 변경
cleand=df.astype({'양치시간(분)':'int','하루양치질(회)':'int','키(cm)':'int','몸무게(kg)':'int'})
cleand.head()
```

```python
#진단일자 데이터 타입 변경 및 년월일시로 칼럼 쪼개기 
cleand['진단일자']= pd.to_datetime(cleand['진단일자'])  
cleand['진단년'] = cleand['진단일자'].dt.year
cleand['진단월'] = cleand['진단일자'].dt.month
cleand['진단일'] = cleand['진단일자'].dt.day
cleand['진단시'] = cleand['진단일자'].dt.hour
cleand['진단요일'] = cleand['진단일자'].dt.weekday
cleand = cleand.drop(columns = ['진단일자'])
```

## 3. Feature Engineering

'y' 하나당 각 1점으로 더해서 점수컬럼생성

1. 흡연, 탄산음료섭취, 양치시간 1분이하, 1회이하 -> 나쁜 습관
2. 사랑니, 비뚠이 -> 바뀔 수 없는 구강구조 
3. 구취, 치아 흔들림, 시린니, 잇몸부종 => 치아 상황 
4. 년,월,일,시, 요일 컬럼 생성

```python
def habit(df,cleand):
    #정수타입으로 변환 전 데이터프레임 사용
    bad_df = df[['흡연','탄산음료섭취','양치시간(분)','하루양치질(회)']]
    bad_df['양치시간(분)']=bad_df['양치시간(분)'].map({'1':'1','2':'0','3':'0'})
    bad_df['하루양치질(회)']=bad_df['하루양치질(회)'].map({'1':'1','2':'0','3':'0'})

    #타입변환
    bad_df=bad_df.astype({'양치시간(분)':'int','하루양치질(회)':'int'})
    bad_df['나쁜습관'] = bad_df.sum(axis=1)

    str_df = df[['사랑니','비뚠니']]
    str_df['바뀔 수 없는 구강구조'] = str_df.sum(axis=1)

    sit_df = df[['구취','치아흔들림','시린니','잇몸부종,통증']]
    sit_df['치아 상황'] = sit_df.sum(axis=1)

    cleand['나쁜습관'] = bad_df['나쁜습관']
    cleand['바뀔 수 없는 구강구조'] = str_df['바뀔 수 없는 구강구조']
    cleand['치아 상황'] = sit_df['치아 상황']

    return cleand
```

```python
habit(df,cleand)
pd.set_option('display.max_columns', 35)
cleand
```
